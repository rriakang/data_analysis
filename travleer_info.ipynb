🎓 학력 분포

**EDU**
1:무학
2:초등학교
3:중학교
4:고등학교
5:대학(4년제미만)
6:대학교(4년제 이상)
7:대학원 석사과정
8:대학원 박사과정

```python
edu_data = data
```

```python
# 학력 코드와 레이블 매핑
edu_labels = {
    1: '무학',
    2: '초등학교',
    3: '중학교',
    4: '고등학교',
    5: '대학(4년제미만)',
    6: '대학교(4년제 이상)',
    7: '대학원 석사과정',
    8: '대학원 박사과정'
}

# 학력 빈도 계산 및 오름차순 정렬
edu_counts = edu_data['EDU_NM'].value_counts().sort_index()  # 학력 코드 기준으로 정렬

# 그래프 스타일 설정
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')  # 스타일 설정
edu_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# x축 레이블을 학력 코드 대신 학력 레이블로 설정
plt.xticks(ticks=range(len(edu_counts)), labels=[edu_labels[key] for key in edu_counts.index], rotation=45, ha='right')

# 그래프 제목과 레이블 설정
plt.title('최종 학력 분포', fontsize=16, fontweight='bold')
plt.xlabel('학력', fontsize=12)
plt.ylabel('빈도', fontsize=12)
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
# 소득 코드와 레이블 매핑
income_labels = {
    1: '소득없음',
    2: '월평균 100만원 미만',
    3: '월평균 100만원 ~ 200만원 미만',
    4: '월평균 200만원 ~ 300만원 미만',
    5: '월평균 300만원 ~ 400만원 미만',
    6: '월평균 400만원 ~ 500만원 미만',
    7: '월평균 500만원 ~ 600만원 미만',
    8: '월평균 600만원 ~ 700만원 미만',
    9: '월평균 700만원 ~ 800만원 미만',
    10: '월평균 800만원 ~ 900만원 미만',
    11: '월평균 900만원 ~ 1,000만원 미만',
    12: '월평균 1,000만원 이상'
}

# 문자열 데이터를 정수형으로 변환
data['HOUSE_INCOME'] = pd.to_numeric(data['HOUSE_INCOME'], errors='coerce')

# 소득 데이터를 레이블로 변환
data['HOUSE_INCOME_LABEL'] = data['HOUSE_INCOME'].map(income_labels)

# 소득 빈도 계산 및 오름차순 정렬
income_counts = data['HOUSE_INCOME_LABEL'].value_counts(ascending=True)

# 그래프 스타일 설정
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')
income_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# 그래프 제목과 레이블 설정
plt.title('가구 소득 분포', fontsize=16, fontweight='bold')
plt.xlabel('소득', fontsize=12)
plt.ylabel('빈도', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# 그래프 출력
plt.show()
```
